```python
import os
import pandas as pd
```

```python


folder_path = '/content/drive/MyDrive/FAERS 212/faers_2024Q4/'

# load in only the drug data files
drug_files = [f for f in os.listdir(folder_path)
              if f.upper().startswith("DRUG") and f.upper().endswith(".TXT")]

# drug_files
```

```python
# Collect all drugname entries into a list
all_drugnames = []

for fname in drug_files:
    fpath = os.path.join(folder_path, fname)
    try:
        # only load in drugname col in FAERS datafiles
        df = pd.read_csv(fpath, sep='$', encoding='latin1', dtype=str, usecols=['drugname'])
        all_drugnames.extend(df['drugname'].dropna().tolist())
    except Exception as e:
        print(f"Failed to load {fname}: {e}")
```

```python
# all_drugnames
```

```python

drugnames_df = pd.DataFrame({'drugname': all_drugnames})
drugnames_df
```

```python
drugnames_df = drugnames_df.drop_duplicates()
```

```python
drugnames_df.to_csv("all_drugnames.csv", index=False)
```
